```python
import os
import glob
import json
import numpy
```

```python
glob.glob("coco_data/train2017/*")
```

```python
inst_train_path = "coco_data/annotations/instances_train2017.json"
inst_train  = json.loads(open(inst_train_path).read())
```

```python
inst_train['annotations'][0]
```

```python
image_id_to_cat = {}
for annotation in inst_train['annotations']:
    cat_id = annotation['category_id']
    if cat_id >= 72 and cat_id <= 77:
        image_id_to_cat[annotation['image_id']] = cat_id
```

```python
from shutil import copyfile

for image in inst_train['images']:
    if image['id'] in image_id_to_cat.keys():
        copyfile("coco_data/train2017/" + image['file_name'], "coco_data/electronics/" + image['file_name'])
```
